```python
# Importing the Libraries
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

# Adjusting the configuration of the plots to be created
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

# Reading the Dataset
movies = pd.read_csv(r'\Users\GIME\Documents\PYTHON PROJECTS\movies.csv')
```

```python
# Checking the Data
movies.head()
```

```python
# Checking for missing values
for col in movies.columns:
    pet_missing = np.mean(movies[col].isnull())
    print('{} - {}%'.format(col, pet_missing))
```

```python
# Removing missing values
movies = movies.dropna()
```

```python
# Rechecking for missing values
for col in movies.columns:
    pet_missing = np.mean(movies[col].isnull())
    print('{} - {}%'.format(col, pet_missing))
```

```python
# Checking the Datatypes
movies.dtypes
```

```python
# Changing Datatype of columns
movies['runtime'] = movies['runtime'].astype('int64')

movies['budget'] = movies['budget'].astype('int64')

movies['votes'] = movies['votes'].astype('int64')

movies['gross'] = movies['gross'].astype('int64')
```

```python
# Rechecking if changes took effect
movies.head()
```

```python
#Rechecking the Datatypes
movies.dtypes
```

```python
# Extracting the correct year of release column
movies['year'] = movies['released'].str.extract(pat = '([0-9]{4})').astype(int)
movies.head()
```

```python
# Sorting the data by the Gross column
movies.sort_values(by=['gross'], inplace=False, ascending=False).head()
```

```python
# Removing duplicates in the data
movies.drop_duplicates()
```

```python
# Using visuals to show the correlation between budget and gross
# Starting with scatterplots
plt.scatter(x=movies['budget'], y=movies['gross'])
plt.title('Correlation between budget and gross earnings')
plt.xlabel('budget')
plt.ylabel('gross earnings')
plt.show()

# Now using seaborn
sns.regplot(x='budget', y='gross', data=movies, scatter_kws={'color':'black'}, line_kws={'color':'indigo'})
plt.show()
```

```python
# Checking the correlation between numeric columns using different methods
movies.corr(numeric_only=True)
# movies.corr(method = 'pearson', numeric_only=True)
# movies.corr(method = 'kendall', numeric_only=True)
# movies.corr(method = 'spearman', numeric_only=True)
```

There is a high correlation between budget and gross

```python
# Visualizing the correlation matrix for the numeric columns
correlation_matrix2 = movies.corr(method = 'pearson', numeric_only=True)
sns.heatmap(correlation_matrix2, annot=True)
plt.title('Correlation matrix for numeric columns')
plt.show()
```

```python
# Showing the correlation between all columns
# First, change the non-numeric columns to numeric columns
movies_numerized = movies
for col_name in movies_numerized.columns:
    if(movies_numerized[col_name].dtype == 'object'):
        movies_numerized[col_name] = movies_numerized[col_name].astype('category')
        movies_numerized[col_name] = movies_numerized[col_name].cat.codes

movies_numerized.head()
```

```python
movies.head()
```

```python
# Then, Visualizing the correlation matrix for all columns
correlation_matrix2 = movies_numerized.corr(method = 'pearson', numeric_only=True)
sns.heatmap(correlation_matrix2, annot=True)
plt.title('Correlation matrix for all columns')
plt.show()
```

```python
# Checking the correlation between all columns using different methods
movies_numerized.corr(numeric_only=True)
# movies.corr(method = 'pearson', numeric_only=True)
# movies.corr(method = 'kendall', numeric_only=True)
# movies.corr(method = 'spearman', numeric_only=True)
```

```python
# For organization, unstacking the ouput above
correlation_matrix = movies_numerized.corr(numeric_only=True)
corr_pairs = correlation_matrix.unstack()

# Display all correlation pairs in sorted order
pd.set_option('display.max_rows', None)
sorted_corr = corr_pairs.sort_values()
sorted_corr
```

```python
# Showing high correlation (>5) between all columns
high_corr = sorted_corr[(sorted_corr) > 0.5]
high_corr
```

Votes and budgets have the highest correlation to gross earnings.
